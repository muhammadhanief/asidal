```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import date_format, count
import matplotlib.pyplot as plt
import calendar

# Konversi 'dt_pos_utc' ke tipe timestamp
df_data = df_data.withColumn("dt_pos_utc", df_data["dt_pos_utc"].cast("timestamp"))

# Ekstrak nama bulan dari timestamp
df_data_dengan_bulan = df_data.withColumn("bulan", date_format("dt_pos_utc", "MMMM"))

# Kelompokkan berdasarkan bulan dan hitung jumlah rekaman
rekaman_per_bulan = df_data_dengan_bulan.groupBy("bulan").agg(count("*").alias("jumlah_record_per_bulan"))

# Daftar nama bulan secara manual (urutkan sesuai keinginan Anda)
bulan_manual = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Kumpulkan hasil perhitungan ke dalam Pandas DataFrame untuk memudahkan visualisasi
record_pd = rekaman_per_bulan.toPandas()

# Ubah nama bulan ke format numerik agar bisa diurutkan dengan benar
record_pd['bulan_numerik'] = record_pd['bulan'].apply(lambda x: bulan_manual.index(x))

# Urutkan DataFrame berdasarkan bulan numerik
record_pd = record_pd.sort_values(by='bulan_numerik')

# Visualisasi
# Tampilkan grafik garis menggunakan matplotlib
plt.plot(record_pd['bulan'], record_pd['jumlah_record_per_bulan'], marker='o')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Record per Bulan')
plt.xticks(rotation=45, ha="right")
plt.ticklabel_format(style='plain', axis='y')
plt.title('Grafik Jumlah Record per Bulan')
plt.show()
```

## Missing Value

```python
import matplotlib.pyplot as plt
from pyspark.sql.functions import col, date_format, sum

# Filter data untuk hanya bulan Januari dan Februari
df_data_filtered = df_data.filter((date_format("dt_pos_utc", "MMMM") == "January") | (date_format("dt_pos_utc", "MMMM") == "February"))

# Hitung jumlah nilai yang hilang per bulan
missing_values_per_month = df_data_filtered.withColumn("months", date_format("dt_pos_utc", "MMMM")) \
    .groupBy("months").agg(*(sum(col(c).isNull().cast("int")).alias(c) for c in df_data.columns))

# Ambil nama bulan dari kolom "months"
months = sorted([month[0] for month in missing_values_per_month.select("months").distinct().rdd.collect()])

# Ambil nilai-nilai yang hilang untuk setiap bulan dan simpan dalam list
values_per_month = []
for month in months:
    values_for_month = missing_values_per_month.filter(col("months") == month).collect()[0]
    values_per_month.append(values_for_month[1:])  # Exclude months from the row

# Transpose list agar sesuai dengan format yang diharapkan oleh plt.plot
values_per_month_transposed = list(map(list, zip(*values_per_month)))

# Tampilkan line chart
plt.figure(figsize=(12, 6))
for i, column in enumerate(df_data.columns[1:]):  # Exclude months from the columns
    plt.plot(months, values_per_month_transposed[i], marker='o', label=column)

plt.xlabel('Bulan')
plt.ylabel('Jumlah Nilai Hilang')
plt.title('Jumlah Nilai Hilang per Bulan (Januari & Februari)')
plt.xticks(rotation=45, ha="right")
plt.legend()
plt.tight_layout()
plt.show()
```

## MMSI Unik

```python
import matplotlib.pyplot as plt
from pyspark.sql.functions import countDistinct, col, date_format

# Hitung jumlah 'mmsi' yang unik per bulan
unique_mmsi_per_month = df_data.withColumn("months", date_format("dt_pos_utc", "MMMM")) \
    .groupBy("months").agg(countDistinct("mmsi").alias("unique_mmsi_count"))

# Ambil nama bulan dari kolom "months"
months = unique_mmsi_per_month.select("months").distinct().rdd.map(lambda x: x[0]).collect()
# Urutkan bulan sesuai dengan urutan kalender
months_sorted = sorted(months, key=lambda x: list(calendar.month_name).index(x.capitalize()))

# Ambil nilai jumlah 'mmsi' yang unik per bulan
unique_mmsi_counts = unique_mmsi_per_month.select("unique_mmsi_count").rdd.map(lambda x: x[0]).collect()

# Tampilkan line chart
plt.figure(figsize=(12, 6))
plt.plot(months_sorted, unique_mmsi_counts, marker='o', color='green')
plt.xlabel('Bulan')
plt.ylabel('Jumlah MMSI Unik')
plt.title('Jumlah MMSI Unik per Bulan')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

## Status Navigasi

```python
import matplotlib.pyplot as plt
from pyspark.sql.functions import countDistinct, col

# Hitung jumlah 'mmsi' yang unik berdasarkan 'nav_status'
unique_mmsi_per_nav_status = df_data.groupBy("nav_status").agg(countDistinct("mmsi").alias("unique_mmsi_count"))

# Ambil nama 'nav_status' dan nilai jumlah 'mmsi' yang unik
nav_status_values = unique_mmsi_per_nav_status.select("nav_status").rdd.map(lambda x: x[0]).collect()
unique_mmsi_counts = unique_mmsi_per_nav_status.select("unique_mmsi_count").rdd.map(lambda x: x[0]).collect()

# Tampilkan bar chart
plt.figure(figsize=(12, 6))
plt.bar(nav_status_values, unique_mmsi_counts, color='blue')
plt.xlabel('Nav Status')
plt.ylabel('Jumlah MMSI Unik')
plt.title('Jumlah MMSI Unik per Nav Status')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

## Fungsi

```python
from IPython.display import HTML
import base64

def create_download_link(df, title = "Download CSV file", filename):
    csv = df.to_csv(index = False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload, title=title, filename=filename)
    return HTML(html)
```
