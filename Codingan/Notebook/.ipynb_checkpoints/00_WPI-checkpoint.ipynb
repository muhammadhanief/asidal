```python
import pandas as pd
import holoviews as hv
from holoviews import opts, dim

hv.extension('matplotlib')
hv.output(fig='svg', size=500)
```

```python
import os
os.chdir('C:/Users/Lenovo/Skripsi/Chord Diagram')

port_df = pd.read_csv('KunjunganPort.csv')
```

```python
ports = list(set(port_df["Port_Prev"].unique().tolist() + port_df["Port_Next"].unique().tolist()))
ports_dataset = hv.Dataset(pd.DataFrame(ports, columns=["Port"]))
```

```python
%%opts Chord [title="Vessel Movement Between Ports" labels="Port"]
%%opts Chord (node_color="Port" node_cmap="Category20" edge_color="Port_Prev" edge_cmap='Category20')

hv.Chord((port_df, ports_dataset))
```
