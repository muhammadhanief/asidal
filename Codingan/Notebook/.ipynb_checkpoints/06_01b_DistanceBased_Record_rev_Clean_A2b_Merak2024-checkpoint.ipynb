## Download csv

```python
def create_download_link(df, title, filename):
    csv = df.to_csv(index = False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload, title=title, filename=filename)
    return HTML(html)
```

## Rata-Rata Draught Masuk

```python
# Hitung rata-rata waktu di pelabuhan
avg_draught_vessel_month = result_in.withColumn("months", F.date_format("dt_pos_utc", "MMMM")) \
    .groupBy("months").agg(F.avg("draught").alias("average_draught_vessel"))
```

```python
# Tampilkan hasil
avg_draught_vessel_month.show()
```

```python
# Hitung rata-rata waktu di pelabuhan menurut tipe kapal
avg_draught_vessel_country = result_in.select("fc_vessel","draught").groupBy("fc_vessel").agg(F.avg("draught").alias("average_draught_vessel"))
```

```python
# Tampilkan hasil
avg_draught_vessel_country.show(avg_draught_vessel_country.count(), truncate = False)
```

```python
# Hitung rata-rata waktu di pelabuhan menurut tipe kapal
avg_draugth_vess_type = result_in.select("vessel_type","draught").groupBy("vessel_type").agg(F.avg("draught").alias("average_draught_vessel"))
```

```python
# Tampilkan hasil
avg_draugth_vess_type.show()
```

## Median Draught Masuk

```python
median_draught_vessel_month = result_in.withColumn("months", F.date_format("dt_pos_utc", "MMMM")) \
    .groupBy("months") \
    .agg(F.expr("percentile_approx(draught, 0.5)").alias("median_draught_vessel"))
```
